```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#importing datasets
```

```python
data =pd.read_csv("TRAIN.csv")
data1=pd.read_csv('TEST.csv')
```

```python
data.head()
```

```python
data.describe().transpose()
```

```python
#Converting into Data into Date and Time
```

```python
import re 
data['Timenew']=re.sub('[^0-1]', '', str(data['Time']))
```

```python
data.head()
```

```python
data1['Timenew']=re.sub('[^0-1]', '', str(data1['Time']))
```

```python
data.head()
```

```python
x=data[['UNIXTime','Temperature','Pressure','Humidity','WindDirection(Degrees)','Speed']]
y=data[['Radiation']]
```

```python
x_test=data1[['UNIXTime','Temperature','Pressure','Humidity','WindDirection(Degrees)','Speed']]
```

```python
#Creating train and validation datasets
```

```python
from sklearn.model_selection import train_test_split
xtrain,xvalid,ytrain,yvalid=train_test_split(x,y,test_size=0.3,random_state=1)
```

Polynomial features allow the use of simpler modeling algorithms as some of the complexity of interpreting the input variables and their relationships is pushed back to the data preparation stage.

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)
xtrain=pf.fit_transform(xtrain)
xvalid=pf.fit_transform(xvalid)
x_test=pf.fit_transform(x_test)
```

```python
#Predicting with Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python

ypred=lr.predict(xvalid)
```

```python
from sklearn.metrics import mean_squared_error as mse
mse(yvalid,ypred)
```

```python
np.sqrt(mse(yvalid,ypred))
```

After using Linear Regression we got Mean Squared error of 180.7487

```python
#Predicting on the test dataset
yfinal=lr.predict(x_test)
yfinal=np.abs(yfinal)
yfinal
```

```python
yfinal=pd.DataFrame(yfinal)
yfinal.to_csv("output1.csv")
```

```python
#Predicting with Random Forest Regressor
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(max_depth=110,max_features=9)
rfr.fit(xtrain,ytrain)
```

```python
y_pred=rfr.predict(xvalid)
```

```python
mse(yvalid,y_pred)
```

```python
np.sqrt(mse(yvalid,y_pred))
```

After using Random Forest Regressor we got Mean Squared error of 117.5204

```python
y_final=rfr.predict(x_test)
print(y_final)
y_final=pd.DataFrame(y_final)
```

```python
y_final.to_csv("output.csv")
```
